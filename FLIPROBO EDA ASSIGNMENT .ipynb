```python
# removing outliers 

from scipy.stats import zscore

z=np.abs(zscore(x))
filtering=(z<3).all(axis=1)
df_validation=df[filtering]
df_validation.describe()
```

```python
df_validation
```

```python
# applying standard scaller 
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(df_validation)
```

```python
scaler_data=scaler.transform(df_validation)
```

```python
scaler_data
```

```python
# apply pca to reduce the size of data 
from sklearn.decomposition import PCA

pca=PCA(n_components=5)
pca.fit(scaler_data)
x_pca=pca.transform(scaler_data)
scaler_data.shape
```
